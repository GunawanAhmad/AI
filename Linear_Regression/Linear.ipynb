```python
import numpy as np
import matplotlib.pyplot as plt 
import random
```

## Create dataset

```python
x_train = np.array([random.randint(1,100) for i in range(100)])
y_train = np.array([x_train[i] + random.randint(-10,10) for i in range(100)])
plt.scatter(x_train, y_train)
plt.show()
```

## Linear Regression

```python
def linear_function(gradient, x):
    y = gradient * x
    return y
```

```python
x_line = np.linspace(0,100,100)
m_init = 3
y_line = linear_function(x_line, m_init)

```

```python
plt.scatter(x_train, y_train)
plt.plot(x_line, y_line, "Red")
plt.show()
```

# training

```python
m = 5
learning_rate = 0.1
for i in range(y_train.size):
    y_prediction = linear_function(m, x_train[i])
    error = y_train[i] - y_prediction
    delta_m = error/x_train[i] * learning_rate
    m = m + delta_m
m
```

```python
plt.scatter(x_train, y_train)
y_prediction = linear_function(x_line, m)
plt.plot(x_line, y_prediction, "red")
plt.show()
```
